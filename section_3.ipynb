```python
img = cv2.imread('Data/00-puppy.jpg')
type(img)
```

```python
plt.imshow(img)
```

```python
img_fixed = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_fixed)
```

```python
img_gray = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
print(img_gray.shape)
plt.imshow(img_gray, cmap='gray')
```

```python
print(img_fixed.shape)
resized_img = cv2.resize(img_fixed, (200, 100))
print(resized_img.shape)
plt.imshow(resized_img)
```

```python
print(img_fixed.shape)
w_ratio = 0.5
h_ratio = 0.5
resized_img2 = cv2.resize(img_fixed, (0, 0), img_fixed, w_ratio, h_ratio) # resize by ratio
print(resized_img2.shape)
plt.imshow(resized_img2)
```

```python
flip = cv2.flip(img_fixed, 0)
plt.imshow(flip)
```
