```python
import cv2

cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(f"width: {width}, height: {height}")

# WINDOWS -- *'DIVX'
# MACOS, LINUX *'XVID'
writer = cv2.VideoWriter('mysupervideo.mp4', cv2.VideoWriter_fourcc(*'DIVX'), 20, (width, height))

while True:
    ret, frame = cap.read()

    # operations (drawing)
    writer.write(frame)

    # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # cv2.imshow('frame', gray)
    cv2.imshow('frame', frame) # direct camera footage

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
writer.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time

cap = cv2.VideoCapture('mysupervideo.mp4')

if cap.isOpened() == False:
    print('Error file not found or wrong codec used!')

while cap.isOpened():
    ret, frame = cap.read()

    if ret == True:
        time.sleep(1/20) # 20 is the frame rate
        cv2.imshow('frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

## top left corner
x = width // 2
y = height // 2

# width & height of the Rectangle
w = width // 4
h = height // 4

# bottom right x+w, y+h


while True:
    ret, frame = cap.read()

    cv2.rectangle(frame, (x, y), (x+w, y+h), color=(0, 0, 255), thickness=4)
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

## CALLBACK FUNCTION RECTANGLE
def draw_rectable(event, x, y, flags, params):
    global pt1, pt2, topLef_clicked, botRight_clicked

    if event == cv2.EVENT_LBUTTONDOWN:

        if topLef_clicked == True and botRight_clicked == True:
            pt1 = (0, 0)
            pt2 = (0, 0)
            topLef_clicked = False
            botRight_clicked = False

        if topLef_clicked == False:
            pt1 = (x, y)
            topLef_clicked = True
        elif botRight_clicked == False:
            pt2 = (x, y)
            botRight_clicked = True


## GLOBAL VARIABLES
pt1 = (0, 0)
pt2 = (0, 0)
topLef_clicked = False
botRight_clicked = False

## CONNECT TO THE CALLBACK
cap = cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test', draw_rectable)


while True:
    ret, frame = cap.read()

    ## DRAWING ON THE FRAME BASED OFF THE GLOBAL VARIABLE
    if topLef_clicked:
        cv2.circle(frame, center=pt1, radius=5, color=(0, 0, 255), thickness=-1)

    if botRight_clicked:
        cv2.rectangle(frame, pt1, pt2, color=(0, 0, 255), thickness=3)

    cv2.imshow('Test', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
